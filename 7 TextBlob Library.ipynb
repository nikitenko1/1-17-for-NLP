<p style="font-family:consolas; font-size: 26px; color: magenta; text-decoration-line: overline; "> Tokenization</p>

```python
'''
Tokenization refers to splitting a large paragraph into sentences or words
'''
from textblob import TextBlob
```

```python
document = ("In computer science, artificial intelligence (AI), \
            sometimes called machine intelligence, is intelligence \
            demonstrated by machines, in contrast to the natural intelligence \
            displayed by humans and animals. Computer science defines AI \
            research as the study of \"intelligent agents\": any device that \
            perceives its environment and takes actions that maximize its\
            chance of successfully achieving its goals.[1] Colloquially,\
            the term \"artificial intelligence\" is used to describe machines\
            that mimic \"cognitive\" functions that humans associate with other\
            human minds, such as \"learning\" and \"problem solving\".[2]")
```

```python
text_blob_object = TextBlob(document)
```

```python
'''Now to get the tokenized sentences, we can use the sentences attribute:'''
document_sentence = text_blob_object.sentences

print(document_sentence)
print(len(document_sentence))
```

```python
'''Similarly, the words attribute returns the tokenized words in the document.'''
document_words = text_blob_object.words

print(document_words)
print(len(document_words))
```

<p style="font-family:consolas; font-size: 26px; color: magenta; text-decoration-line: overline; "> Lemmatization</p>

```python
'''Lemmatization refers to reducing the word to its root form as found in a dictionary.'''
'''you have to use the Word object from the textblob library, 
pass it the word that you want to lemmatize and then call the lemmatize method'''
from textblob import Word

word1 = Word("apples")
print("apples:", word1.lemmatize())

word2 = Word("media")
print("media:", word2.lemmatize())

word3 = Word("greater")
print("greater:", word3.lemmatize("a")) # "a" as a parameter This specifically tells the method that the word should be treated as an adjective
```

<p style="font-family:consolas; font-size: 26px; color: magenta; text-decoration-line: overline; "> Parts of Speech (POS) Tagging</p>

```python
'''To find POS tags for the words in a document, all you have to do is use the tags attribute'''
for word, pos in text_blob_object.tags:
    print(word + " => " + pos)
```

<p style="font-family:consolas; font-size: 26px; color: magenta; text-decoration-line: overline; "> Convert Text to Singular and Plural</p>

```python
'''to convert text into a plural or singular form using the pluralize and singularize methods'''
text = ("Football is a good game. It has many health benefit")
text_blob_object = TextBlob(text)
print(text_blob_object.words.pluralize())
```

```python
'''Similarly, to singularize words you can use singularize method as follows:'''
text = ("Footballs is a goods games. Its has many healths benefits")

text_blob_object = TextBlob(text)
print(text_blob_object.words.singularize())
```

<p style="font-family:consolas; font-size: 26px; color: magenta; text-decoration-line: overline; "> Noun Phrase Extraction</p>

```python
'''To find noun phrases, you simply have to use the noun_phrase attributes on the TextBlob object'''
text_blob_object = TextBlob(document)
for noun_phrase in text_blob_object.noun_phrases:
    print(noun_phrase)
```

<p style="font-family:consolas; font-size: 26px; color: magenta; text-decoration-line: overline; "> Getting Words and Phrase Counts</p>

```python
'''To find the frequency of occurrence of a particular word'''
text_blob_object = TextBlob(document)
text_blob_object.word_counts['intelligence']
```

```python
'''It is important to mention that by default the search is not case-sensitive'''
text_blob_object.words.count('intelligence')
```

```python
'''search to be case sensitive, you need to pass True as the value for the case_sensitive parameter'''
text_blob_object.words.count('intelligence', case_sensitive=True)
```

```python
text_blob_object = TextBlob(document)
text_blob_object.noun_phrases.count('artificial intelligence')
```

<p style="font-family:consolas; font-size: 26px; color: magenta; text-decoration-line: overline; "> Converting to Upper and Lowercase</p>

```python
text = "I love to watch football, but I have never played it"
text_blob_object = TextBlob(text)

print(text_blob_object.upper())
```

```python
text = "I LOVE TO WATCH FOOTBALL, BUT I HAVE NEVER PLAYED IT"
text_blob_object = TextBlob(text)

print(text_blob_object.lower())
```

<p style="font-family:consolas; font-size: 26px; color: magenta; text-decoration-line: overline; "> Finding N-Grams</p>

```python
'''For instance, for a sentence "I love watching football", 
some 2-grams would be (I love), (love watching) and (watching football). 
N-Grams can play a crucial role in text classification'''
text = "I love to watch football, but I have never played it"
text_blob_object = TextBlob(text)
for ngram in text_blob_object.ngrams(2):
    print(ngram)
```

<p style="font-family:consolas; font-size: 26px; color: magenta; text-decoration-line: overline; "> Spelling Corrections</p>

```python
'''With the correct method of the TextBlob object, you can correct all the spelling mistakes in your text'''
text = "I love to watchf footbal, but I have neter played it"
text_blob_object = TextBlob(text)

print(text_blob_object.correct())
```

<p style="font-family:consolas; font-size: 26px; color: magenta; text-decoration-line: overline; "> Text Classification</p>

```python
'''TextBlob also provides basic text classification capabilities. '''
train_data = [
    ('This is an excellent movie', 'pos'),
    ('The move was fantastic I like it', 'pos'),
    ('You should watch it, it is brilliant', 'pos'),
    ('Exceptionally good', 'pos'),
    ("Wonderfully directed and executed. I like it", 'pos'),
    ('It was very boring', 'neg'),
    ('I did not like the movie', 'neg'),
    ("The movie was horrible", 'neg'),
    ('I will not recommend', 'neg'),
    ('The acting is pathetic', 'neg')
]
test_data = [
    ('Its a fantastic series', 'pos'),
    ('Never watched such a brillent movie', 'pos'),
    ("horrible acting", 'neg'),
    ("It is a Wonderful movie", 'pos'),
    ('waste of money', 'neg'),
    ("pathetic picture", 'neg')
]
```

```python
'''To train the model, we simply have to pass the training data to the constructor of the NaiveBayesClassifier class'''
from textblob.classifiers import NaiveBayesClassifier

classifier = NaiveBayesClassifier(train_data)
```

```python
print(classifier.classify("It is very boring"))
'''It looks like a negative review. When you execute the above script, you will see neg in the output'''
```

```python
print(classifier.classify("It's a fantastic series"))
'''It looks like a negative review. When you execute the above script, you will see neg in the output'''
```

```python
classifier.accuracy(test_data)
```

```python
'''To find the most important features for the classification, the show_informative_features method can be used'''
classifier.show_informative_features(3)
```

<p style="font-family:consolas; font-size: 26px; color: magenta; text-decoration-line: overline; "> Sentiment Analysis</p>

```python
import pandas as pd
import numpy as np

reviews_datasets = pd.read_csv('./data/Reviews.csv')
reviews_datasets = reviews_datasets.head(20000)
reviews_datasets.dropna()
reviews_datasets.sample(2)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
 
sns.histplot(reviews_datasets['Score'])
plt.show()
```

```python
'''You can see that most of the ratings are highly positive i.e. '''
 
sns.countplot(x='Score', data=reviews_datasets)
plt.show()
```

```python
reviews_datasets['Text'][350]
```

```python
'''Let's find the sentiment of the 350th review.'''
text_blob_object = TextBlob(reviews_datasets['Text'][350])
print(text_blob_object.sentiment)
```

```python
'''Let's now add a column for sentiment polarity in our dataset:'''
def find_pol(review):
    return TextBlob(review).sentiment.polarity

reviews_datasets['Sentiment_Polarity'] = reviews_datasets['Text'].apply(find_pol)
reviews_datasets.head()
```

```python
sns.distplot(reviews_datasets['Sentiment_Polarity'])
plt.show()
```

```python
sns.barplot(x='Score', y='Sentiment_Polarity', data=reviews_datasets)
plt.show()
```

```python
most_negative = reviews_datasets[reviews_datasets.Sentiment_Polarity == -1].Text.head()
print(most_negative)
```

```python
'''Let's now see some of the most positive reviews:'''
most_positive = reviews_datasets[reviews_datasets.Sentiment_Polarity == 1].Text.head()
print(most_positive)
```
